```python
import pandas as pd
```

```python
from sqlalchemy import create_engine
```

```python
csv_file = "./Resources/indian_food.csv"
indian_food_data_df = pd.read_csv(csv_file)
indian_food_data_df.head()
```

```python
indian_food_data_df['first_ingredient'] = indian_food_data_df["ingredients"].str.split(',').str[0]
indian_food_data_df
indian_food_data_df['second_ingredient'] = indian_food_data_df["ingredients"].str.split(',').str[1]
indian_food_data_df
indian_food_data_df['third_ingredient'] = indian_food_data_df["ingredients"].str.split(',').str[2]
indian_food_data_df
indian_food_data_df['fourth_ingredient'] = indian_food_data_df["ingredients"].str.split(',').str[3]
indian_food_data_df
```

```python
indian_food_data_df = indian_food_data_df.dropna(how='any')
indian_food_data_df
```

```python
clean_india_food_df = indian_food_data_df[['ingredients', 'first_ingredient', 'second_ingredient', 'third_ingredient', 'fourth_ingredient']]
clean_india_food_df

```

```python
clean_india_food_df
clean_india_food_df.to_csv('clean_india_food.csv')
```

```python
nutrition_csv_file = "./Resources/nutrition.csv"
nutrition_data_df = pd.read_csv(nutrition_csv_file)
nutrition_data_df.head()
```

```python
nutrition_data_df = nutrition_data_df.rename(columns={"name": "ingredients"})
nutrition_data_df
```

```python
nutrition_data_df['first_ingredient'] = nutrition_data_df["ingredients"].str.split(',').str[0]
nutrition_data_df
nutrition_data_df['second_ingredient'] = nutrition_data_df["ingredients"].str.split(',').str[1]
nutrition_data_df
```

```python
nutrition_data_df = nutrition_data_df.dropna(how='any')
nutrition_data_df
```

```python
clean_nutrition_df = nutrition_data_df[['ingredients', 'first_ingredient', 'second_ingredient']]
clean_nutrition_df
```

```python
connection_string = "postgres:270504@localhost:5432/ingredients_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
clean_india_food_df.to_sql(name='india_food', con=engine, if_exists='append', index=False)
```

```python
clean_nutrition_df.to_sql(name='nutrition_data', con=engine, if_exists='append', index=False)
```
